```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = df.drop('koi_disposition',axis=1) # use all features
y = df.loc[:,'koi_disposition']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1) # split
```

```python
X_train.head()
```

```python
y_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Import Scaler and train it
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# Create scaled test and train
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Run a random forest to determine feature importance

```python
# Create a random forest classifier and score
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_test_scaled, y_test)
```

```python
# Determine Importances
importances = rf.feature_importances_
importance_df = pd.DataFrame(sorted(zip(rf.feature_importances_, X.columns), reverse=True))
importance_df
```
